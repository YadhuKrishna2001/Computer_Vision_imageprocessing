Import Libraries

```python
import cv2
import numpy as np
sfrom google.colab.patches import cv2_imshow
```

Import Image File


 Read the image


```python
image = cv2.imread('/content/OIP.jpg')
```

 Resize the image


```python
resized_width = 300
resized_height = 300
resized = cv2.resize(image, (resized_width, resized_height))
print(resized)
cv2_imshow(resized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Convert to grayscale


```python
gray = cv2.cvtColor(resized, cv2.COLOR_BGR2GRAY)
print(gray)
cv2_imshow(gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

 Noise reduction


```python
denoised = cv2.fastNlMeansDenoising(gray, None, h=10)
print(denoised)
cv2_imshow(denoised)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

 Apply Gaussian blur


```python
blurred = cv2.GaussianBlur(denoised, (5, 5), 0)
print(blurred)
cv2_imshow(blurred)
cv2.waitKey(0)
cv2.destroyAllWindows()

```

Adjust contrast (optional)


```python
alpha = 1.5  # Contrast control (1.0-3.0)
adjusted = cv2.convertScaleAbs(blurred, alpha=alpha, beta=0)
cv2_imshow(adjusted)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Save or display the preprocessed image
cv2.imwrite('new_oip.jpg', adjusted)
```

Edge detecting


```python
threshold1 = 225
threshold2 = 225
edge =cv2.Canny(image,threshold1,threshold2)
print(edge)
cv2_imshow(image)
cv2_imshow(edge)
```
